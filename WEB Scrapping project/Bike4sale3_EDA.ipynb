# üèçÔ∏è **Univariate Analysis:**

## CATEGORICAL COLUMNS

### Brand
###  1. Count of top 10 most common brands 

```python
plt.figure(figsize=(10,6))
brand_counts = df['Brand'].value_counts().head(10)

ax=sns.barplot(x=brand_counts.values, y=brand_counts.index, palette="viridis")
for i in ax.containers:
    ax.bar_label(i)
plt.title("Top 10 Most Common Bike Brands Listed for Sale", fontsize=14)
plt.xlabel("Number of Listings")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()
```

- **Bajaj** has the highest number of listings, followed by **Royal Enfield** and **Honda**.

- **Hero**, **TVS**, and **Yamaha** also show strong resale presence.

- **Indian** brands dominate the used bike market overall.

### Bike Model
### 2. Which bike model is most selled

```python
model = df['Bike Model'].value_counts().head(10)
```

```python
plt.figure(figsize =(12,4))
ax = sns.barplot(y = model.index, x = model.values, hue = model, palette='viridis',legend = False)

for i in ax.containers:
    ax.bar_label(i)

plt.ylabel("Bike Model")
plt.xlabel("Number of bikes")
plt.title('Most Bike Model Sell')
plt.savefig("Most_Bike_Model_sell.png", dpi=400, bbox_inches="tight")
plt.show()
```

1. **Royal Enfield Classic 350** is the top-selling model, showing strong brand popularity in the used bike market.  
2. **Royal Enfield Thunderbird 350** ranks second, confirming high demand for Royal Enfield bikes overall.  
3. **Ola S1 Pro** (an electric scooter) appears in the top 3, showing the growing interest in EVs.  
4. Common commuter bikes like **Honda Activa** and **Bajaj Pulsar 150 DTSi** also maintain steady resale demand.  
5. Overall, both **Royal Enfield** and **Ola Electric** dominate the most listed models segment.

### Model Year
