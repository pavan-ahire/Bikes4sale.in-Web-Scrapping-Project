```python
def get_model_type(model):
    model = str(model).lower()
    if any(x in model for x in ['r15', 'duke', 'rc', 'speed', 'ninja', 'gs', 'rr', 'rs', 'race', 'yzf', 'apache rr']):
        return 'Sports'
    elif any(x in model for x in ['activa', 'jupiter', 'ntorq', 'pleasure', 'maestro', 'burgman', 'fascino', 'scooty', 'sxl', 'vxl', 'ray zr']):
        return 'Scooter'
    elif any(x in model for x in ['classic', 'meteor', 'bullet', 'avenger', 'roadster', 'bobber', 'imperiale', 'cruise']):
        return 'Cruiser'
    elif any(x in model for x in ['xpulse', 'himalayan', 'adventure', 'scram', 'ronin', 'v-strom']):
        return 'Adventure'
    elif any(x in model for x in ['electric', 's1', 'revolt', 'iqube', 'optima', 'ev', 'etron', 'battery', 'magnus', 'photon', 'komaki', 'faast']):
        return 'Electric'
    else:
        return 'Commuter'

df['Model_Type'] = df['Bike Model'].apply(get_model_type)
```
